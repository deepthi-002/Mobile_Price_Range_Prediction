```python
X = df.drop('price_range',axis=1)
y = df['price_range']
```

Train-Test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred1 = model1.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print("The Accuracy Score is",accuracy_score(y_pred1,y_test))
print("\n")
print("The Classification report is",classification_report(y_pred1,y_test))
print("\n")
print("The Confusion matrix is",confusion_matrix(y_pred1,y_test))
```

```python
from sklearn.svm import SVC
model2 = SVC(kernel='rbf')
model2.fit(X_train,y_train)
```

```python
y_pred2 = model2.predict(X_test)
```

```python
y_pred2
```

```python
print("The Accuracy Score is",accuracy_score(y_pred2,y_test))
print("\n")
print("The Classification report is",classification_report(y_pred2,y_test))
print("\n")
print("The Confusion matrix is",confusion_matrix(y_pred2,y_test))
```

```python
model2 = SVC(kernel='linear')
```

```python
model2.fit(X_train,y_train)
```

```python
y_pred3 = model2.predict(X_test)
```

```python
y_pred3
```

```python
print("The Accuracy Score is",accuracy_score(y_pred3,y_test))
print("\n")
print("The Classification report is",classification_report(y_pred3,y_test))
print("\n")
print("The Confusion matrix is",confusion_matrix(y_pred3,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=15,metric='euclidean') 
```

```python
model3.fit(X_train,y_train)
```

```python
y_pred4 = model3.predict(X_test) 
```

```python
y_pred4
```

```python
print("The Accuracy Score is",accuracy_score(y_pred4,y_test))
print("\n")
print("The Classification report is",classification_report(y_pred4,y_test))
print("\n")
print("The Confusion matrix is",confusion_matrix(y_pred4,y_test))
```
